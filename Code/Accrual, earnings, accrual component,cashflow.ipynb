```python
import pandas as pd
```

```python
netIncomeFromContinuingOps=pd.read_csv('netIncomeFromContinuingOps.csv')
bse_netIncomeFromContinuingOps=pd.read_csv('bse_netIncomeFromContinuingOps.csv')
netIncome=pd.read_csv('netIncome.csv')
bse_netIncome=pd.read_csv('bse_netIncome.csv')
totalCashFromOperatingActivities=pd.read_csv('totalCashFromOperatingActivities.csv')
bse_totalCashFromOperatingActivities=pd.read_csv('bse_totalCashFromOperatingActivities.csv')
average_assets=pd.read_csv('average_assets.csv')
bse_average_assets=pd.read_csv('bse_average_assets.csv')


netIncomeFromContinuingOps.rename(columns = {'Unnamed: 0':'Company'}, inplace = True) 
bse_netIncomeFromContinuingOps.rename(columns = {'Unnamed: 0':'Company'}, inplace = True) 
netIncome.rename(columns = {'Unnamed: 0':'Company'}, inplace = True) 
bse_netIncome.rename(columns = {'Unnamed: 0':'Company'}, inplace = True) 
totalCashFromOperatingActivities.rename(columns = {'Unnamed: 0':'Company'}, inplace = True) 
bse_totalCashFromOperatingActivities.rename(columns = {'Unnamed: 0':'Company'}, inplace = True) 
average_assets.rename(columns = {'Unnamed: 0':'Company'}, inplace = True) 
bse_average_assets.rename(columns = {'Unnamed: 0':'Company'}, inplace = True) 
```

```python
#formula used: Income from Continuing Operations / Average Total Assets 
nse_Earning=pd.DataFrame(columns=netIncomeFromContinuingOps.columns)
nse_Earning['Company']=netIncomeFromContinuingOps['Company']
nse_Earning['2020-03-31']=netIncomeFromContinuingOps['2020-03-31']/average_assets['2020-03-31']
nse_Earning['2019-03-31']=netIncomeFromContinuingOps['2019-03-31']/average_assets['2019-03-31']
nse_Earning['2018-03-31']=netIncomeFromContinuingOps['2018-03-31']/average_assets['2018-03-31']
nse_Earning['2017-03-31']=netIncomeFromContinuingOps['2017-03-31']/average_assets['2017-03-31']
nse_Earning

bse_Earning=pd.DataFrame(columns=bse_netIncomeFromContinuingOps.columns)
bse_Earning['Company']=bse_netIncomeFromContinuingOps['Company']
bse_Earning['2020-03-31']=bse_netIncomeFromContinuingOps['2020-03-31']/bse_average_assets['2020-03-31']
bse_Earning['2019-03-31']=bse_netIncomeFromContinuingOps['2019-03-31']/bse_average_assets['2019-03-31']
bse_Earning['2018-03-31']=bse_netIncomeFromContinuingOps['2018-03-31']/bse_average_assets['2018-03-31']
bse_Earning['2017-03-31']=bse_netIncomeFromContinuingOps['2017-03-31']/bse_average_assets['2017-03-31']
bse_Earning

#Merging nse and bse earnings to get total earnings 

Earnings=pd.concat([nse_Earning,bse_Earning])
Earnings

#Cleaned accruals with no NaN values
Earnings.dropna(inplace=True)
Earnings
```

```python
#Formula used:Total Accrual= Net profit - Net Cash from Operating Activities
nse_Accruals=pd.DataFrame(columns=netIncome.columns)
nse_Accruals.Company=netIncome.Company
nse_Accruals['2020-03-31']=netIncome['2020-03-31']-totalCashFromOperatingActivities['2020-03-31']
nse_Accruals['2019-03-31']=netIncome['2019-03-31']-totalCashFromOperatingActivities['2019-03-31']
nse_Accruals['2018-03-31']=netIncome['2018-03-31']-totalCashFromOperatingActivities['2018-03-31']
nse_Accruals['2017-03-31']=netIncome['2017-03-31']-totalCashFromOperatingActivities['2017-03-31']
nse_Accruals

bse_Accruals=pd.DataFrame(columns=bse_netIncome.columns)
bse_Accruals.Company=bse_netIncome.Company
bse_Accruals['2020-03-31']=bse_netIncome['2020-03-31']-bse_totalCashFromOperatingActivities['2020-03-31']
bse_Accruals['2019-03-31']=bse_netIncome['2019-03-31']-bse_totalCashFromOperatingActivities['2019-03-31']
bse_Accruals['2018-03-31']=bse_netIncome['2018-03-31']-bse_totalCashFromOperatingActivities['2018-03-31']
bse_Accruals['2017-03-31']=bse_netIncome['2017-03-31']-bse_totalCashFromOperatingActivities['2017-03-31']
bse_Accruals

#Merging nse and bse accrual to get total accruals 

Accruals=pd.concat([nse_Accruals,bse_Accruals])
Accruals

is_NaN = Accruals.isnull()
row_has_NaN = is_NaN.any(axis=1)
Company_with_NaN_data_Accruals = Accruals[row_has_NaN]
Company_with_NaN_data_Accruals

#Cleaned accruals with no NaN values
Accruals.dropna(inplace=True)
Accruals
```

```python
#formula used:Accruals/ Average Total Assets 
nse_Accruals_Accrual_Component=pd.DataFrame(columns=nse_Accruals.columns)
nse_Accruals_Accrual_Component.Company=nse_Accruals.Company
nse_Accruals_Accrual_Component['2020-03-31']=nse_Accruals['2020-03-31']/average_assets['2020-03-31']
nse_Accruals_Accrual_Component['2019-03-31']=nse_Accruals['2019-03-31']/average_assets['2019-03-31']
nse_Accruals_Accrual_Component['2018-03-31']=nse_Accruals['2018-03-31']/average_assets['2018-03-31']
nse_Accruals_Accrual_Component['2017-03-31']=nse_Accruals['2017-03-31']/average_assets['2017-03-31']
nse_Accruals_Accrual_Component

bse_Accruals_Accrual_Component=pd.DataFrame(columns=nse_Accruals.columns)
bse_Accruals_Accrual_Component.Company=nse_Accruals.Company
bse_Accruals_Accrual_Component['2020-03-31']=bse_Accruals['2020-03-31']/bse_average_assets['2020-03-31']
bse_Accruals_Accrual_Component['2019-03-31']=bse_Accruals['2019-03-31']/bse_average_assets['2019-03-31']
bse_Accruals_Accrual_Component['2018-03-31']=bse_Accruals['2018-03-31']/bse_average_assets['2018-03-31']
bse_Accruals_Accrual_Component['2017-03-31']=bse_Accruals['2017-03-31']/bse_average_assets['2017-03-31']
bse_Accruals_Accrual_Component

#Merging nse and bse accrual components to get accrual component

Accrual_component=pd.concat([nse_Accruals_Accrual_Component,bse_Accruals_Accrual_Component])
Accrual_component

#Cleaned accrual component with no NaN values
Accrual_component.dropna(inplace=True)
Accrual_component
```

```python
#Cashflow component formula used: Income from Continuing Operations -Accruals / Average Total Assets 
nse_Cashflow_component=pd.DataFrame(columns=nse_Accruals.columns)
nse_Cashflow_component.Company=nse_Accruals.Company
nse_Cashflow_component['2020-03-31']=netIncomeFromContinuingOps['2020-03-31']-nse_Accruals['2020-03-31']/average_assets['2020-03-31']
nse_Cashflow_component['2019-03-31']=netIncomeFromContinuingOps['2019-03-31']-nse_Accruals['2019-03-31']/average_assets['2020-03-31']
nse_Cashflow_component['2018-03-31']=netIncomeFromContinuingOps['2018-03-31']-nse_Accruals['2018-03-31']/average_assets['2020-03-31']
nse_Cashflow_component['2017-03-31']=netIncomeFromContinuingOps['2017-03-31']-nse_Accruals['2017-03-31']/average_assets['2020-03-31']

bse_Cashflow_component=pd.DataFrame(columns=nse_Accruals.columns)
bse_Cashflow_component.Company=nse_Accruals.Company
bse_Cashflow_component['2020-03-31']=bse_netIncomeFromContinuingOps['2020-03-31']-bse_Accruals['2020-03-31']/bse_average_assets['2020-03-31']
bse_Cashflow_component['2019-03-31']=bse_netIncomeFromContinuingOps['2019-03-31']-bse_Accruals['2019-03-31']/bse_average_assets['2020-03-31']
bse_Cashflow_component['2018-03-31']=bse_netIncomeFromContinuingOps['2018-03-31']-bse_Accruals['2018-03-31']/bse_average_assets['2020-03-31']
bse_Cashflow_component['2017-03-31']=bse_netIncomeFromContinuingOps['2017-03-31']-bse_Accruals['2017-03-31']/bse_average_assets['2020-03-31']
bse_Cashflow_component

#Merging nse and bse cashflow components to get cashflow component

Cashflow_component=pd.concat([nse_Cashflow_component,bse_Cashflow_component])
Cashflow_component

#Cleaned cashflow component with no NaN values
Cashflow_component.dropna(inplace=True)
Cashflow_component
```
